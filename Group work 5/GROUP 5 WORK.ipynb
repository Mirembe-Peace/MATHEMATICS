You have been provided a dataset labelled "Loan_Approval_Data.csv" which has information regarding loan's applications in a commercial bank.

Use the dataset to answer the following questions:

What is the marginal probability of any continuous variable? [7 MARKS]
What is the marginal probability of any categorical variable? [5 MARKS]
What is the probability that a "female" applicant has her Loan_Status reflecting "Y?" [9 MARKS]
Explain the central tendency of any two continuous variables. [5 MARKS]
Explain the relationship between the variables "Gender" and "Loan_Status." [5 MARKS]

```python
import pandas as pd
```

```python
# loading the dataset
file = 'Loan_Approval_Data.csv'
loan_data = pd.read_csv(file)
```

```python
# Question 1: What is the marginal probability of any continous variable?
applicant_income_density = loan_data['ApplicantIncome'].plot.kde()
applicant_income_density.set_title('Marginal ProbabilityDensity of ApplicantIncome')

# Peaks in the curve show where the income values/ranges are most common among the applicants.
```

```python
# Question 2: What is the marginal probability of any categorical variable?
education_counts = loan_data['Education'].value_counts(normalize=True)
print("Marginal Probability of Education Levels:")
print(education_counts)

# The probabilities for the 'Education' levels reflect a portion of each level in the dataset i.e. amongst graduates and non graduates
```

```python
# Question 3: What is the probability that a 'female' applicant has her
Female_applicants = loan_data[loan_data['Gender'] == 'Female']
prob_females_loan_approved = Female_applicants['Loan_Status'].value_counts(normalize=True).get('Y',0)
print(f"Probability ofloanapproval for female applicants: {prob_females_loan_approved}")

# the value computed is the portion of female applicants who got their loans approved.
```

```python
# Question 4: Explain the central tendency of any two continous variables
applicant_income_mean = loan_data['ApplicantIncome'].mean()
applicant_income_median = loan_data['ApplicantIncome'].median()
applicant_income_mode = loan_data['ApplicantIncome'].mode()[0]
print(f"ApplicantIncome - Mean: {applicant_income_mean}, Median: {applicant_income_median}, Mode: {applicant_income_mode}")
# these measures give an overall sense of the income distribution in the applicants

loan_amount_mean = loan_data['LoanAmount'].mean()
loan_amount_median = loan_data['LoanAmount'].median()
loan_amount_mode = loan_data['LoanAmount'].mode()[0]
print(f"LoanAmount - Mean: {loan_amount_mean}, Median: {loan_amount_median}, Mode: {loan_amount_mode}")
# these values give an understanding of which loan amounts are requested for by the applicants
```

```python
# Question 5: Explain the relationship between the variables "Gender" and "Loan_Status"
gender_loan_status = pd.crosstab(loan_data['Gender'], loan_data['Loan_Status'], normalize='index')
print("Relationship between Gender and Loan_Status:")
print(gender_loan_status)

# this shows the percentage of loans approved and denied for female and male applicants respectively.
```
